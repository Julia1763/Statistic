```python
#Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 
#    75, 65, 84, 90, 150. 
#Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, 
#среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

import numpy as np
import math 

money_arr=np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
# Среднее арифметическое:
X = sum(money_arr)/len(money_arr)
print(X)
```

```python
# среднее квадратичное отклонение:
square_dev=(money_arr-X)**2
SKO = (sum(square_dev)/len(money_arr))**(1/2)
print(SKO)
```

```python
# смещенную оценку дисперсий:
square_dev=(money_arr-X)**2
sig = sum(square_dev)/len(money_arr)
print(sig)
```

```python
# несмещенную оценку дисперсий:
square_dev=(money_arr-X)**2
sig_2 = sum(square_dev)/(len(money_arr)-1)
print(sig_2)
```

```python
# В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. 
# Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. 
# Какова вероятность того, что 3 мяча белые?
from math import factorial
def combinations(n, k):
    return factorial(n) / (factorial(k) * factorial(n - k))

# Найдем вероятность, что все 3 белых мяча вытащили из второй корзины:
P1 = (combinations(3,2) / combinations(8,2)) * (combinations(5,3) * combinations(7,1) / combinations(12,4))
print(P1)
# Найдем вероятность, что 2 белых мяча вытащили из второй корзины, а 1 мяч - из первой:
P2 = (combinations(5,1) * combinations(3,1) * combinations(5,2) * combinations(7,2)) / (combinations(8,2) * combinations(12,4))
print(P2)
# Найдем вероятность, что 1 белый мяч вытащили из второй корзины, а 2 мяча - из первой:
P3 = (combinations(5,2) * combinations(5,1) * combinations(7,3)) / (combinations(8,2) * combinations(12,4))
print(P3)

P = P1 + P2 + P3
print(P)
```

```python
# На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. 
# Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. 
# Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом 
#    в). третьим спортсменом.
# Найдем полную вероятность:
P1 = 0.9
P2 = 0.8
P3 = 0.6
P = 1/3 * (P1 + P2 + P3)
# Найдем вероятность, что выстрел произведен 1м спортсменом
P_01 = (1/3 * P1) / P
# Найдем вероятность, что выстрел произведен 2м спортсменом
P_02 = (1/3 * P2) / P
# Найдем вероятность, что выстрел произведен 3м спортсменом
P_03 = (1/3 * P3) / P
print(P_01)
print(P_02)
print(P_03)
```

```python
# В университет на факультеты A и B поступило равное количество студентов, 
# а на факультет C студентов поступило столько же, сколько на A и B вместе. 
# Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. 
# Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. 
# Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B 
#    в). на факультете C?
# Найдем полную вероятность:
P1 = 0.8
P2 = 0.7
P3 = 0.9
# 1/x + 1/x + 2/x = 1 => x = 4
P_0 = 1/4 * (0.8 + 0.7) + 1/2 * 0.9
# Найдем вероятность, что студент учится на факультете А
P_01 = ((1/4) * P1) / P_0
# Найдем вероятность, что студент учится на факультете B
P_02 = ((1/4) * P2) / P_0
# Найдем вероятность, что студент учится на факультете C
P_03 = ((1/2) * P3) / P_0
print(P_01)
print(P_02)
print(P_03)
```

```python
# Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, 
# для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: 
#    а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?
# Все детали вышли из строя:
P1 = 0.1
P2 = 0.2
P3 = 0.25
P_001 = P1 * P2 * P3
# Вышли только две детали:
P_002 = P1 * P2 * (1 - P3) + P1 * P3 * (1 - P2) + P2 * P3 * (1 - P1)
# Не вышла ни одна деталь:
P_003 = 1 - P_001
# Вышла одна деталь:
P_004 = P1 * (1 - P2) * (1 - P3) + P2 * (1 - P1) * (1 - P2) + P3 * (1 - P3) * (1 - P1)
print(P_001)
print(P_002)
print(P_003)
print(P_004)
```
